```python
def straight_line_distance(state):
    heuristic_values = {
        "Arad": 366, 
        "Zerind": 374, 
        "Timisoara": 329, 
        "Sibiu": 253,
        "Oradea": 380,
        "Lugoj": 244, 
        "Fagaras": 176, 
        "Rimnicu Vilcea": 193, 
        "Mehadia": 241,
        "Drobeta": 242,
        "Craiova": 160, 
        "Pitesti": 100, 
        "Bucharest": 0, 
        "Giurgiu": 77,
        "Eforie" : 161,
        "Hirsova" : 151,
        "Iasi" : 226,
        "Neamt" : 234,
        "Urziceni" : 80,
        "Vaslui" : 199
    }
    return heuristic_values[state]
```

```python
class Node:
    def __init__(self, state, child, f = 0, g = 0,father = None):
        self.father = father
        self.state = state
        self.child = child
        self.g = g
        self.f = f
```

```python
def expand_map(state):
    return graph[state]
```

```python
def UniformCostSearch(initial, target, successor_function):

    target_path = target
    node = Node(state=initial, child=successor_function(initial))
    frontier = []
    explored_states = []
    explored_nodes = []
    pathway = []

    frontier.append(node)


    while frontier:
        
        min = frontier[0].g
        index_min = 0

        for i in range(len(frontier)):
            if frontier[i].g < min:
                min = frontier[i].g
                index_min = i
        
        print("\nFronteira:")
        for n in frontier:
            print(f"({n.g}, {n.state}),", end=' ')
        print()
    
        print(f"\nO menor custo é de: {(frontier[index_min].g, frontier[index_min].state)}")
        first_node = frontier.pop(index_min)
        explored_nodes.append(first_node)

        if first_node.state == target_path:
            for n in reversed(explored_nodes):
                if n.state == target_path:
                    pathway.append(n.state)
                    target_path = n.father
            pathway.reverse()
            print("Chegou ao destino!")
            return f"Caminho: {pathway}, Custo: {first_node.g}, Expansões: {len(explored_nodes) - 1}"
        
        explored_states.append(first_node.state)
        print(f"Lista de explorados: {explored_states}")
        print(f"{first_node.state} abre para: ")

        for state in first_node.child:

            print(state)
            node = Node(state = state[1], child = successor_function(state[1]), g = first_node.g + state[0], father = first_node.state)
            
            frontier_names = [frontier_node.state for frontier_node in frontier]
            
            
            if node.state not in explored_states and node.state not in frontier_names:
                frontier.append(node)
            elif node.state in frontier_names:
                same_state = [n for n in frontier if n.state == node.state]
                if node.g < same_state[0].g:
                    print(f"Remove -> {same_state[0].g}")
                    print(f"Adiciona -> {node.g}")
                    frontier.remove(same_state[0])
                    frontier.append(node)
```

```python
start_time = perf_counter()
result_ucs = UniformCostSearch(initial="Arad", target="Bucharest", successor_function=expand_map)
end_time = perf_counter()

print("Resultado:", result_ucs)
print("Tempo de execução:", end_time - start_time, "segundos")
```

```python
def IterativeDeepeningSearch(initial, target, successor_function):
    i = 0
    res = None
    while res == None:
        print(f"----------------- Iteração {i} -----------------", end='')
        res = DepthLimitedSearch(initial=initial, target=target, limit=i, successor_function = successor_function)
        if type(res) == tuple:
            return f"Caminho: {res[0]},Custo: {res[1]}, Expansões: {len(res[2]) - 1}"
        i +=1
    
def DepthLimitedSearch(initial, target, successor_function, limit):
    node = Node(state=initial, child=successor_function(initial), g = 0, father=None)
    return RecursiveDLS(target=target, node=node, successor_function=successor_function, limit=limit, pathway = [], explored_nodes= [])
        
       
def RecursiveDLS(target, node, limit, successor_function, pathway, explored_nodes):
    

    print("\nFronteira:")
    print(f"({node.g}, {node.state}),")
    
    explored_nodes.append(node)
    pathway.append(node.state)

    print("\nNós explorados:")
    for en in explored_nodes:
        print(f"({en.g}, {en.state}),", end=' ')
    print()

    
    if node.state == target:
        return (pathway, node.g, explored_nodes)
    elif limit == 0:
        print("Limite alcançado")
        return None
    else:
        print(f"{node.state} abre para: ")
        print(f"{node.child}")
        for state in node.child:
            child_node = Node(state=state[1], child=successor_function(state[1]), g = node.g + state[0], father = node.state)
            result = RecursiveDLS(target=target, limit=limit-1, node=child_node, successor_function = successor_function, pathway = pathway.copy(), explored_nodes=explored_nodes)
            if type(result) == tuple:
                return result
```

```python
start_time = perf_counter()
result_ids = IterativeDeepeningSearch(initial="Arad", target="Bucharest", successor_function=expand_map)
end_time = perf_counter()


print("Resultado:", result_ids)
print("Tempo de execução:", end_time - start_time, "segundos")
```
